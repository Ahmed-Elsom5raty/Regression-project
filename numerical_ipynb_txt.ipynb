```python
import numpy as np
import matplotlib.pyplot as plt

class NumericalRegression:
    def __init__(self, learning_rate=0.01, iterations=1000, lasso=0.0, ridge=0.0):
        self.lr = learning_rate
        self.iterations = iterations
        self.lasso = lasso
        self.ridge = ridge
        self.m = 0
        self.b = 0
        self.loss = []

    def fit(self, X, y):
        X = np.array(X).flatten()
        y = np.array(y).flatten()
        n = len(X)
        for _ in range(self.iterations):
            y_pred = self.m * X + self.b
            mse = np.mean((y - y_pred) ** 2)
            reg = 0
            if self.ridge > 0:
                reg = self.ridge * (self.m ** 2)
            elif self.lasso > 0:
                reg = self.lasso * abs(self.m)
            total_loss = mse + reg
            self.loss.append(total_loss)
            dm = (-2/n) * np.sum(X * (y - y_pred))
            db = (-2/n) * np.sum(y - y_pred)
            if self.ridge > 0:
                dm += 2 * self.ridge * self.m
            elif self.lasso > 0:
                dm += self.lasso * np.sign(self.m)

            self.m -= self.lr * dm
            self.b -= self.lr * db

    def predict(self, X):
        X = np.array(X).flatten()
        return self.m * X + self.b

    def evaluate(self, X, y):
        y_pred = self.predict(X)
        return np.mean((y - y_pred) ** 2)

    def plot_loss(self):
        plt.plot(self.loss)
        plt.xlabel("Iterations")
        plt.ylabel("Loss")
        plt.title("Loss over Iterations")
        plt.show()

    def get_parameters(self):
        return self.m, self.b
```

```python
df = pd.read_csv("Students_Grades_Predictions.csv")


X = df["StudyHoursPerWeek"].fillna(df["StudyHoursPerWeek"].mean())
y = df["FinalGrade"].fillna(df['FinalGrade'].mean())



model = NumericalRegression(
    learning_rate=0.001,
    iterations=1000,
    ridge=0.1
    )


model.fit(X, y)

model.plot_loss()

predictions = model.predict(X)


mse = model.evaluate(X, y)


m, b = model.get_parameters()

print("Slope (m):", m)
print("Intercept (b):", b)
print("MSE:", mse)
```
