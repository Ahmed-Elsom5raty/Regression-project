## Linear Regression in Statistical view 

We will use these equations to implement our Model :

- $\hat{\beta} = (X^T X)^{-1} X^T y $
- $\hat{Y}= X \hat{\beta}$
- $R^2=1-\frac{SSE}{SST}$
- $Adjusted \enspace R^2 =1 − (1 − R^2)\frac{n−k} {n−1}$ where $k$ : Number of estimators
  
Our source is [this video](https://youtu.be/NzuK4iAfxhU?si=aenkBw8LLyzo-5Bc) I suggest you to watch the video for deeper understanding

```python
import numpy as np
import pandas as pd
```

```python
class statisticalRegression: 
    def __init__(self):
        self.beta=None # 2-d array of betas including beta node 
        self.mse=None
        self.sse=None
        self.sst=None
        self.r_squared=None
        self.adjusted_r_squared=None
        self.p=None
        
    def fit(self,y,x):
        Y=y.to_numpy()
        XwithoutIntercept=x.to_numpy()
        n= XwithoutIntercept.shape[0]
        ones = np.ones(n)
        X=np.column_stack([ones, XwithoutIntercept]) # To predict beta_0 we included its coeffient (1)
        self.p= X.shape[1] #number of attributes including beta node
        self.beta=np.linalg.inv(X.T @ X) @ X.T @ y # @ is multiplication operator in Matrices
        Y_hat=X @ self.beta #nxP @ Px1 ==> Yhat is nx1
        e=Y-Y_hat
        self.sse=np.sum(e**2)
        self.sst = np.sum((Y - np.mean(Y))**2)
        self.mse=self.sse/(n-self.p) 
        self.r_squared= 1-(self.sse/self.sst) 
        self.adjusted_r_squared=1-(1-self.r_squared) * (n-1)/(n-self.p)
        
    def predict(self,x):
        XwithoutIntercept=x.to_numpy()
        n= XwithoutIntercept.shape[0]
        ones = np.ones(n)
        X=np.column_stack([ones, XwithoutIntercept]) # To predict beta_0 we incluse its coeffient (1)
        predictedY= X @ self.beta
        return predictedY
        
    def summary(self):
        print(f"R squared = {self.r_squared}")
        print(f"Adjusted R squared = {self.adjusted_r_squared}")
        print(f"Numer of predictors = {self.p}")
```
